Data Analysis

```python
# 1- number of orders according to date

df['Date'] = pd.to_datetime(df['Date'])  # Convert 'Date' column to datetime format
orders_by_date = df.groupby(df['Date'].dt.date)['Order ID'].count()
print(orders_by_date)
plt.figure(figsize=(14, 7))
plt.plot(orders_by_date.index, orders_by_date.values, marker='o',markerfacecolor='black', linestyle='-', color='violet')
plt.title('Number of Orders by Date', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Orders', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 2- Order Amount Analysis
print("\nOrder Amount Analysis:")
print("Total Amount: ", df['Amount'].sum())
print("Average Order Amount: ", df['Amount'].mean())
print("Maximum Order Amount: ", df['Amount'].max())
print("Minimum Order Amount: ", df['Amount'].min())
plt.figure(figsize=(10, 6))
plt.hist(df['Amount'], bins=30, color='blue', edgecolor='red')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title('Order Amount Distribution')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3- Top 5 Sales by Category
sales_by_category = df.groupby('Category')['Order ID'].count().sort_values(ascending=False)
print(sales_by_category)
plt.figure(figsize=(10, 5))
bars = plt.barh(sales_by_category.index[:5], sales_by_category.values[:5], color=['red','coral','lightblue','cyan']) 
plt.title(' Top 5 Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales Amount')
plt.bar_label(bars)
plt.show()
```

```python
# 4- sales by size
sales_by_size = df.groupby('Size')['Order ID'].count().sort_values(ascending=False)
print(sales_by_size)
plt.figure(figsize=(12, 8))
bars = plt.bar(sales_by_size.index, sales_by_size.values, color='mediumpurple')
plt.title('Sales by Size', fontsize=16)
plt.xlabel('Size', fontsize=14)
plt.ylabel('Number Of Orders', fontsize=14)
plt.bar_label(bars)
plt.tight_layout()
plt.show()
```

```python
# 5-  Top 5 SKUs by Quantity Sold
top_skus = df.groupby('SKU')['Qty'].sum().nlargest(10)
print(top_skus)
plt.figure(figsize=(10, 5))
bars = plt.bar(top_skus.index[:5], top_skus.values[:5], color=['coral','red','lightblue','cyan','lightgreen']) 
plt.title('Top 5 SKUs by Quantity Sold')
plt.xlabel('SKU')
plt.ylabel('Total Quantity Sold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.bar_label(bars)
plt.show()
```

```python
# 6- B2B vs. B2C sales
b2b_vs_b2c = df.groupby('B2B')['Order ID'].count() # Group by 'B2B' and sum the 'Amount'
print(b2b_vs_b2c)
plt.figure(figsize=(12, 6))
bars = plt.bar(b2b_vs_b2c.index[:10], b2b_vs_b2c.values[:10], width=0.3, color=['lightgreen', 'red'])
plt.title('B2B vs. B2C Sales')
plt.xticks([0, 1], ['B2C', 'B2B'])
plt.xlabel('Business Type')
plt.bar_label(bars)
plt.ylabel('Total Sales')
plt.show()
```

```python
# 7- Top 10 Cities by Total Orders
orders_by_city = df.groupby('ship-city')['Order ID'].count().sort_values(ascending=False)
print(orders_by_city)

plt.figure(figsize=(12, 6))
bars = plt.bar(orders_by_city.index[:10], orders_by_city.values[:10], color='coral') 

plt.title('Top 10 Cities by Total Orders', fontsize=16)
plt.xlabel('City', fontsize=14)
plt.ylabel('Number of Orders', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.bar_label(bars)

plt.show()
```

```python
# 8- Top 10 States by Total Orders
orders_by_state = df.groupby('ship-state')['Order ID'].count().sort_values(ascending=False)
print(orders_by_state)
plt.figure(figsize=(12, 6))
bars = plt.bar(orders_by_state.index[:10], orders_by_state.values[:10], color='lightblue')  # Limiting to top 10 states for clarity
plt.title('Top 10 States by Total Orders', fontsize=16)
plt.xlabel('State', fontsize=14)
plt.ylabel('Number of Orders', fontsize=14)
plt.bar_label(bars)
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Show the plot
plt.tight_layout()
plt.show()
```

```python
# 9- Status  of  Total orders
# Group by 'Status' and calculate the number of orders
orders_by_status = df.groupby('Status')['Order ID'].count()

plt.figure(figsize=(12, 6))
bars = plt.bar(orders_by_status.index, orders_by_status.values, width=0.4,color='red', alpha=0.6)
plt.title('Number of Orders by Status', fontsize=16)
plt.xticks(rotation=90)
plt.xlabel('Order Status')
plt.ylabel('Number of Orders', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.bar_label(bars)
# Show the plot
plt.tight_layout()
plt.show()
```

```python
# 10- Shipping Service Level Distribution
shipping_service_level = df.groupby('ship-service-level')['Order ID'].count()
print(shipping_service_level)
plt.figure(figsize=(12,6))
bars = plt.bar(shipping_service_level.index[:10], shipping_service_level[:10], width=0.4, color=['lightgreen', 'red'])
plt.title('Shipping Service Level Distribution')
plt.xticks([0, 1], ['Standard', 'Expedited'])
plt.xlabel('Shipping Service Level')
plt.bar_label(bars)
plt.ylabel('Total orders')
plt.show()
```
